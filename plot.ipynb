```python
import matplotlib.pyplot as plt

logger_path = 'logs/test_TFN.log'
file = open(logger_path)
epoch_id = []
training_loss,training_acc = [],[]
evaluation_loss,evaluation_acc = [],[]

for row in file:
    if 'Namespace' in row:
        continue

    row = row.split(': ')
    if 'train_epoch_loss' in row[0]:
        training_loss.append(float(row[1]))
    elif 'evalutation_epoch_loss' in row[0]:
        evaluation_loss.append(float(row[1])) 
    elif 'train_epoch_acc' in row[0]:
        training_acc.append(float(row[1])) 
    elif 'evalutation_epoch_acc' in row[0]:
        evaluation_acc.append(float(row[1])) 
    elif 'epoch ' == row[0]:
        epoch_id.append(int(row[1]))
    else:    
        continue


```

```python
plt.plot(epoch_id,training_loss,label='training loss')
plt.plot(epoch_id,evaluation_loss,label='evaluation loss')
plt.legend()
plt.show()
```

```python
plt.plot(epoch_id,training_acc,label='training acc')
plt.plot(epoch_id,evaluation_acc,label='evaluation losaccs')

plt.legend()
plt.show()
```
